## 1. Built-in Functions (Review)

Python comes with many built-in functions. Let's review a few.


```python
print('=' * 50)
print('BUILT-IN FUNCTIONS')
print('=' * 50)

# print() - displays output
print('Hello, World!')

# len() - returns length
numbers = [1, 2, 3, 4, 5]
print(f'Length of list: {len(numbers)}')

# sum() - adds numbers
total = sum(numbers)
print(f'Sum of numbers: {total}')

# max() and min() - find extremes
print(f'Maximum: {max(numbers)}')
print(f'Minimum: {min(numbers)}')

# type() - returns data type
print(f'Type of numbers: {type(numbers)}')
```

## 2. Creating Your First Function

Let's create our own functions using the `def` keyword.


```python
print('\n' + '=' * 50)
print('DEFINING FUNCTIONS')
print('=' * 50)

# Syntax:
# def function_name():
#     code to execute

# Example 1: Simple function with no parameters
def greet():
    print('Hello! Welcome to Python Functions!')

# Calling the function
greet()
print()
```

```python
# Example 2: Function that performs a calculation
def calculate_tax():
    price = 100
    tax_rate = 0.08
    tax = price * tax_rate
    print(f'Tax on ${price}: ${tax}')

calculate_tax()
print()
```

```python
# Example 3: Multiple function calls
def display_separator():
    print('-' * 40)

display_separator()
print('Important Message')
display_separator()
print('Another Message')
display_separator()
```

## 3. Functions with Parameters

Parameters allow you to pass data into functions, making them flexible and reusable.


```python
print('=' * 50)
print('SINGLE PARAMETER')
print('=' * 50)

# Function with one parameter
def greet_person(name):
    print(f'Hello, {name}! Welcome!')

# Calling with different arguments
greet_person('Alice')
greet_person('Bob')
greet_person('Charlie')
print()
```

```python
# Example: Calculate square
def calculate_square(number):
    square = number ** 2
    print(f'The square of {number} is {square}')

calculate_square(5)
calculate_square(10)
calculate_square(7)
print()
```

```python
print('=' * 50)
print('MULTIPLE PARAMETERS')
print('=' * 50)

# Function with multiple parameters
def calculate_rectangle_area(length, width):
    area = length * width
    print(f'Rectangle: {length} × {width} = {area} square units')

calculate_rectangle_area(5, 3)
calculate_rectangle_area(10, 8)
print()
```

```python
# Example: Personalized greeting with multiple parameters
def greet_with_details(name, age, city):
    print(f'Hello {name}!')
    print(f'Age: {age}')
    print(f'City: {city}')
    print('-' * 30)

greet_with_details('Alice', 25, 'New York')
greet_with_details('Bob', 30, 'London')
```

## 4. Return Values

Functions can **return** values that you can store and use later.


```python
print('=' * 50)
print('RETURN VALUES')
print('=' * 50)

# Without return - just prints
def add_numbers_print(a, b):
    result = a + b
    print(f'{a} + {b} = {result}')

add_numbers_print(5, 3)  # Just displays
# Can't use the result elsewhere!

print()
```

```python
# With return - gives back a value
def add_numbers_return(a, b):
    result = a + b
    return result  # Sends the value back

# Now we can store and use the result!
sum1 = add_numbers_return(5, 3)
sum2 = add_numbers_return(10, 20)

print(f'First sum: {sum1}')
print(f'Second sum: {sum2}')
print(f'Total: {sum1 + sum2}')
print()
```

```python
# Example: Calculate discount
def calculate_discount(price, discount_percent):
    discount_amount = price * (discount_percent / 100)
    final_price = price - discount_amount
    return final_price

# Use the returned value
original_price = 100
discounted = calculate_discount(original_price, 20)

print(f'Original: ${original_price}')
print(f'After 20% discount: ${discounted}')
print()
```

```python
# Returning multiple values (as a tuple)
def get_statistics(numbers):
    total = sum(numbers)
    count = len(numbers)
    average = total / count
    return total, count, average  # Returns a tuple

scores = [85, 90, 78, 92, 88]
sum_val, count_val, avg_val = get_statistics(scores)

print(f'Scores: {scores}')
print(f'Sum: {sum_val}')
print(f'Count: {count_val}')
print(f'Average: {avg_val:.2f}')
```

## 5. Default Parameters

Set default values for parameters to make them optional.


```python
print('=' * 50)
print('DEFAULT PARAMETERS')
print('=' * 50)

# Function with default parameter
def greet(name='Guest'):
    print(f'Welcome, {name}!')

# Call with argument - uses provided value
greet('Alice')

# Call without argument - uses default value
greet()

print()
```

```python
# Example: Calculate total with optional tax
def calculate_total(price, tax_rate=0.08):
    """Calculate total price including tax"""
    tax = price * tax_rate
    total = price + tax
    return total

# Using default tax rate (8%)
price1 = 100
total1 = calculate_total(price1)
print(f'${price1} with default tax (8%): ${total1:.2f}')

# Using custom tax rate
price2 = 100
total2 = calculate_total(price2, 0.10)  # 10% tax
print(f'${price2} with custom tax (10%): ${total2:.2f}')

print()
```

```python
# Multiple default parameters
def create_user_profile(name, age=18, country='USA', active=True):
    print(f'User Profile:')
    print(f'  Name: {name}')
    print(f'  Age: {age}')
    print(f'  Country: {country}')
    print(f'  Active: {active}')
    print('-' * 30)

# Using all defaults
create_user_profile('Alice')

# Override some defaults
create_user_profile('Bob', age=25)

# Override with keyword arguments
create_user_profile('Charlie', country='UK', age=30)
```

## 6. Positional vs Keyword Arguments

You can pass arguments by position or by name.


```python
print('=' * 50)
print('POSITIONAL VS KEYWORD ARGUMENTS')
print('=' * 50)

def describe_pet(animal_type, pet_name):
    print(f'I have a {animal_type}.')
    print(f"My {animal_type}'s name is {pet_name}.")
    print()

# Positional arguments (order matters!)
describe_pet('dog', 'Buddy')

# Keyword arguments (order doesn't matter)
describe_pet(pet_name='Whiskers', animal_type='cat')

# Mix of both (positional first, then keyword)
describe_pet('hamster', pet_name='Squeaky')
```

## 7. Variable Scope (Local vs Global)

Understanding where variables can be accessed.


```python
print('=' * 50)
print('VARIABLE SCOPE')
print('=' * 50)

# Global variable (accessible everywhere)
company_name = 'DataCorp'

def display_company():
    # Can access global variable
    print(f'Company: {company_name}')

display_company()
print()
```

```python
# Local variable (only inside function)
def calculate_bonus():
    salary = 50000  # Local variable
    bonus = salary * 0.10
    print(f'Bonus: ${bonus}')

calculate_bonus()

# This would cause an error:
# print(salary)  # salary doesn't exist outside the function!

print()
```

```python
# Local variable shadows global
total = 1000  # Global

def calculate_with_local():
    total = 100  # Local (different from global)
    print(f'Inside function: {total}')

calculate_with_local()
print(f'Outside function: {total}')  # Global unchanged
```

## 8. Practical Data Analytics Examples

Real-world functions for data analysis tasks.


```python
print('\n' + '=' * 50)
print('EXAMPLE 1: TEMPERATURE CONVERTER')
print('=' * 50)

def celsius_to_fahrenheit(celsius):
    """Convert Celsius to Fahrenheit"""
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

def fahrenheit_to_celsius(fahrenheit):
    """Convert Fahrenheit to Celsius"""
    celsius = (fahrenheit - 32) * 5/9
    return celsius

# Test the functions
temp_c = 25
temp_f = celsius_to_fahrenheit(temp_c)
print(f'{temp_c}°C = {temp_f}°F')

temp_f2 = 77
temp_c2 = fahrenheit_to_celsius(temp_f2)
print(f'{temp_f2}°F = {temp_c2:.1f}°C')
```

```python
print('\n' + '=' * 50)
print('EXAMPLE 2: DATA VALIDATION')
print('=' * 50)

def validate_email(email):
    """Check if email is valid"""
    if '@' in email and '.' in email:
        return True
    return False

def validate_age(age):
    """Check if age is valid (18-120)"""
    if 18 <= age <= 120:
        return True
    return False

# Test validation
test_emails = ['user@example.com', 'invalid.email', 'good@test.org']

print('Email Validation:')
for email in test_emails:
    is_valid = validate_email(email)
    status = '✓ Valid' if is_valid else '✗ Invalid'
    print(f'  {email}: {status}')

print('\nAge Validation:')
test_ages = [25, 15, 150, 45]
for age in test_ages:
    is_valid = validate_age(age)
    status = '✓ Valid' if is_valid else '✗ Invalid'
    print(f'  Age {age}: {status}')
```

```python
print('\n' + '=' * 50)
print('EXAMPLE 3: SALES CALCULATIONS')
print('=' * 50)

def calculate_revenue(quantity, price, discount=0):
    """Calculate total revenue with optional discount"""
    subtotal = quantity * price
    discount_amount = subtotal * (discount / 100)
    total = subtotal - discount_amount
    return total

def calculate_commission(sales_amount, rate=0.10):
    """Calculate sales commission"""
    commission = sales_amount * rate
    return commission

# Test sales calculations
qty = 10
unit_price = 50

revenue = calculate_revenue(qty, unit_price)
print(f'Revenue (no discount): ${revenue:.2f}')

revenue_discounted = calculate_revenue(qty, unit_price, discount=15)
print(f'Revenue (15% discount): ${revenue_discounted:.2f}')

commission = calculate_commission(revenue)
print(f'Commission (10%): ${commission:.2f}')

commission_higher = calculate_commission(revenue, rate=0.15)
print(f'Commission (15%): ${commission_higher:.2f}')
```

```python
print('\n' + '=' * 50)
print('EXAMPLE 4: GRADE CALCULATOR')
print('=' * 50)

def calculate_average(scores):
    """Calculate average of scores"""
    total = sum(scores)
    count = len(scores)
    average = total / count
    return average

def get_letter_grade(score):
    """Convert numeric score to letter grade"""
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

def analyze_student(name, test_scores):
    """Analyze student performance"""
    avg = calculate_average(test_scores)
    grade = get_letter_grade(avg)
    
    print(f'Student: {name}')
    print(f'  Scores: {test_scores}')
    print(f'  Average: {avg:.2f}')
    print(f'  Grade: {grade}')
    print()

# Test with students
analyze_student('Alice', [85, 92, 88, 90])
analyze_student('Bob', [78, 82, 75, 80])
analyze_student('Charlie', [95, 98, 92, 96])
```

## Solution 1: Simple Calculator


```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b != 0:
        return a / b
    else:
        return 'Error: Division by zero'

# Test the calculator
print(f'10 + 5 = {add(10, 5)}')
print(f'10 - 5 = {subtract(10, 5)}')
print(f'10 × 5 = {multiply(10, 5)}')
print(f'10 ÷ 5 = {divide(10, 5)}')
print(f'10 ÷ 0 = {divide(10, 0)}')
```

## Solution 2: BMI Calculator


```python
def calculate_bmi(weight, height):
    """Calculate BMI from weight (kg) and height (m)"""
    bmi = weight / (height ** 2)
    return bmi

def categorize_bmi(bmi):
    """Categorize BMI value"""
    if bmi < 18.5:
        return 'Underweight'
    elif bmi < 25:
        return 'Normal'
    elif bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

# Test
weight = 70  # kg
height = 1.75  # meters

bmi = calculate_bmi(weight, height)
category = categorize_bmi(bmi)

print(f'Weight: {weight}kg, Height: {height}m')
print(f'BMI: {bmi:.1f}')
print(f'Category: {category}')
```

## Solution 3: Password Strength Checker


```python
def check_password_strength(password):
    """Check password strength based on length"""
    length = len(password)
    
    if length < 6:
        return 'Weak'
    elif length < 10:
        return 'Medium'
    else:
        return 'Strong'

# Test with multiple passwords
passwords = ['123', 'pass123', 'SecurePassword123']

print('Password Strength Analysis:')
for pwd in passwords:
    strength = check_password_strength(pwd)
    print(f'  "{pwd}" ({len(pwd)} chars): {strength}')
```

## Solution 4: Discount Calculator


```python
def apply_discount(price, discount_percent, min_purchase=0):
    """Apply discount if minimum purchase is met"""
    if price >= min_purchase:
        discount_amount = price * (discount_percent / 100)
        final_price = price - discount_amount
        return final_price
    else:
        return price  # No discount applied

# Test scenarios
print('Discount Scenarios:')

price1 = apply_discount(100, 20)
print(f'$100, 20% off, no min: ${price1:.2f}')

price2 = apply_discount(50, 15, min_purchase=75)
print(f'$50, 15% off, min $75: ${price2:.2f} (no discount)')

price3 = apply_discount(200, 25, min_purchase=100)
print(f'$200, 25% off, min $100: ${price3:.2f}')
```

## Solution 5: List Statistics


```python
def get_list_stats(numbers):
    """Calculate statistics for a list of numbers"""
    total = sum(numbers)
    average = total / len(numbers)
    minimum = min(numbers)
    maximum = max(numbers)
    
    return total, average, minimum, maximum

# Test
numbers = [23, 45, 12, 67, 34, 89, 11, 56]

total, avg, min_val, max_val = get_list_stats(numbers)

print(f'Numbers: {numbers}')
print(f'Sum: {total}')
print(f'Average: {avg:.2f}')
print(f'Minimum: {min_val}')
print(f'Maximum: {max_val}')
```
